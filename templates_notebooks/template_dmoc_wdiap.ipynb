```python
#___LOAD FESOM2 REFERENCE DATA________________________________________________________________________
# load divergence of density class
if ref_path is not None:
    print(ref_path, ref_name)
    data_DIV_ref = tpv.load_dmoc_data(mesh, ref_path, ref_name, year, 'dmoc', std_dens, do_wdiap=True, do_info=False) 

#___LOAD FESOM2 DATA___________________________________________________________________________________    
# load divergence of density class
data_DIV_list = list()
for datapath, descript in zip(input_paths, input_names): 
    print(datapath, descript)
    data_DIV_list.append( tpv.load_dmoc_data(mesh, datapath, descript, year, 'dmoc', std_dens, do_wdiap=True, do_info=False) ) 

#___COMPUTE DIAPYCNAL VERTICAL VELOCITY________________________________________________________________
# finds closest index of isopycnal class
ndens       = len(std_dens)
idx_isopycn = np.argmin(np.abs(np.array(std_dens)-which_isopyc))
print(' sigma_2 = {:5.2f} kg/m^3'.format(std_dens[idx_isopycn]))

# integrate (xr.sum) divergence over all density classes below idx_isopycn
if ref_path is not None:
    wdiap_ref = data_DIV_ref.isel(ndens=range(ndens-1,idx_isopycn-1,-1)).sum(dim='ndens',skipna=True).rename({'std_dens_DIV':'w_diap'})
    wdiap_ref[list(wdiap_ref.keys())[0]] = wdiap_ref[list(wdiap_ref.keys())[0]].assign_attrs({'description':'diapycnal velocity',
                                                                                              'long_name'  :'diapycnal velocity',
                                                                                              'units'      :'m/s',
                                                                                              'str_ldep'   :', $\sigma_{{2}}$={:5.2f}kg/m³'.format(std_dens[idx_isopycn])})

# integrate (xr.sum) divergence over all density classes below idx_isopycn     
wdiap_list = list()
for data in data_DIV_list:
    wdiap = data.isel(ndens=range(ndens-1,idx_isopycn-1,-1)).sum(dim='ndens',skipna=True).rename({'std_dens_DIV':'w_diap'})
    wdiap[list(wdiap.keys())[0]] = wdiap[list(wdiap.keys())[0]].assign_attrs({'description':'diapycnal velocity',
                                                                              'long_name'  :'diapycnal velocity',
                                                                              'units'      :'m/s',
                                                                              'str_ldep'   :', $\sigma_{{2}}$={:5.2f}kg/m³'.format(std_dens[idx_isopycn])})
    
    if ref_path is not None: wdiap_list.append( tpv.do_anomaly(wdiap, wdiap_ref) )
    else:                    wdiap_list.append( wdiap )    
```
